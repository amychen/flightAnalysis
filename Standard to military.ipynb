```python
def convert24(s):
    x = s[-2:]
    y = s[-8:-6]
    if (x == 'am') and (y == '12'):
        ss = ("00" + s[-6:-3]).strip()
    elif x == 'am' and y != '10' and y != '11':
        ss = ('0' + s[:-2]).strip()
    elif x == 'am':
        ss = s[:-2].strip()        
    elif (x == 'pm') and (y == '12'):
        ss = (s[:-2]).strip()
    else:
        ss = (str(int(y) +12) + s[-6:-3]).strip()
    return ss
df['converted'] = df['24time'].apply(convert24)
```
